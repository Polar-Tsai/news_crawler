PTT形式 - bs4


| Title     | Description  |
| ----------|:-------------:|
| Label     | Real = 1, Fake = 0
| Headline  | Headline of News            |
| Link      | url of News |
| Date      | Published Date      |  
| Source    | Source of the news. (One who can verify the claim of the news)| 
| Category  | Category of News            |
| Content   | Article or body of the news |

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
```

```python
# 建立pandas表格
table = {
    "Headline" : [],
    "Link" : [],
    "Date" : [],
    # "Source" : [],
    "Content" : []
}
```

```python
import requests.packages.urllib3 # 解決InsecureRequestWarning
requests.packages.urllib3.disable_warnings()
```

```python
u = 0


for u in range(50):
    u = u + 1
    url = "https://www.ptt01.net/article/newlyArticleList?page="  + str(u + 1)
    print("page:", (u + 1))
    response = requests.get(url, verify = False)
    html = BeautifulSoup(response.text)
    posts = html.find_all("div", class_ = "card")
    for post in posts:
        headline = post.find("strong").text
        link = post.find("a")["href"]
        table["Headline"].append(headline)
        table["Link"].append(link)

        # print(headline)
        # print(date, source, category)
        print(link)

        # 取出link進入取得內文、發布日期
        response_ = requests.get(link, verify = False)
        html_ = BeautifulSoup(response_.text)
        # print(html_)
        date = html_.find("div", class_ = "article-tip").find("small").text
        table["Date"].append(date)
        try:
            contents = html_.find("div", class_ = "article-body").find_all("p")
            li = []
            for c in contents:
                content = c.text.strip()
                li.append(content)
                joinli = "\n".join(li)
            table["Content"].append(joinli)
        except:
            print("=" * 20)
            print("NONE TYPE")
            print("=" * 20)
            table["Content"].append("NONE TYPE")
            continue
        # print(joinli) # .text要放這裡，若放在contents會被視為nonetype
        # print("=" * 20)
```

```python
df = pd.DataFrame(table)
df
```
