PTT形式 - bs4


| Title     | Description  |
| ----------|:-------------:|
| Label     | Real = 1, Fake = 0
| Headline  | Headline of News            |
| Link      | url of News |
| Date      | Published Date      |  
| Source    | Source of the news. (One who can verify the claim of the news)| 
| Category  | Category of News            |
| Content   | Article or body of the news |

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
```

```python
# 建立pandas表格
table = {
    "Headline" : [],
    "Link" : [],
    "Date" : [],
    "Source" : [],
    "Category" : [],
    "Content" : []
}
```

```python
import requests.packages.urllib3 # 解決InsecureRequestWarning
requests.packages.urllib3.disable_warnings()
```

```python
page = 100

for u in range(page):
    url = "https://www.thenewslens.com/category/health?page="  + str(u + 1)
    print("page:", (u + 1))
    response = requests.get(url, verify = False)
    html = BeautifulSoup(response.text)
    posts = html.find_all("div", class_ = "list-container ")
    for post in posts:
        headline = post.find("h2", class_ = "title").text
        date = post.find("h3", class_ = "info-content").find("span", class_ = "time").text
        date = date.strip(" | ")
        link = post.find("a")["href"]
        table["Headline"].append(headline)
        table["Date"].append(date)
        table["Link"].append(link)

        source = post.find("span", class_ = "author").text
        table["Source"].append(source)
        category = url.split("/")[-1].split("?")[0]
        table["Category"].append(category)
        # print(headline)
        # print(date, source, category)
        print(link)

        # 取出link進入取得內文
        response_ = requests.get(link, verify = False)
        html_ = BeautifulSoup(response_.text)
        # print(html_)
        try:
            contents = html_.find("div", class_ = "article-content").find_all("p")
            li = []
            for c in contents:
                content = c.text.strip()
                li.append(content)
                joinli = "\n".join(li)
            table["Content"].append(joinli)
        except:
            print("=" * 20)
            print("NONE TYPE")
            print("=" * 20)
            table["Content"].append("NONE TYPE")
            continue
        # print(joinli) # .text要放這裡，若放在contents會被視為nonetype
        # print("=" * 20)
```
